```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn import decomposition
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
# readind the dataset
df = pd.read_csv('consumer_compliants.csv')
```

```python
complaints_df=df[['Consumer complaint narrative','Product','Company']].rename(columns={'Consumer complaint narrative':'complaints'})
```

```python
pd.set_option('display.max_colwidth',-1)
complaints_df
```

```python
X_train,X_test = train_test_split(complaints_df,test_size=0.6,random_state=111)
```

```python
X_train['Product'].value_counts()
```

```python
lem = WordNetLemmatizer()
```

```python
def tokenize(text):
    tokens = [word for word in nltk.word_tokenize(text) if (len(word) > 3 and len(word.strip('XXxx/')) > 2)]
    lems = [lem.lemmatize(item) for item in tokens]
    lems = list(set(lems))
    return lems
```

```python
vectorizer_tf = TfidfVectorizer(tokenizer=tokenize,stop_words='english',max_df=0.75,min_df=50,max_features=1000,use_idf=False,norm=None)
tf_vectors = vectorizer_tf.fit_transform(X_train.complaints)
```

```python
tf_vectors.A
```

```python
vectorizer_tf.get_feature_names()
```

```python
lda = decomposition.LatentDirichletAllocation(n_components=6,max_iter=3,learning_method='online',learning_offset=50,random_state=111)
W1 = lda.fit_transform(tf_vectors)
H1 = lda.components_
```

```python
W1
```

```python
H1
```

```python
num_words=15

vocab = np.array(vectorizer_tf.get_feature_names())

top_words = lambda t : [vocab[i] for i in np.argsort(t)[:-num_words-1:-1]]
topic_words = ([top_words(t) for t in H1])
topics = [' '.join(t) for t in topic_words]
```

```python
topics
```

```python
colnames = ["Topic_" + str(i) for i in range(lda.n_components)]
colnames
```

```python
docnames = ["Doc_" + str(i) for i in range(len(X_train.complaints))]
docnames[:10]
```

```python
df_doc_topic = pd.DataFrame(np.round(W1,2),columns=colnames,index=docnames)
df_doc_topic.head(10)
```

```python
significant_topic = np.argmax(df_doc_topic.values,axis=1)
significant_topic[:30]
```

```python
df_doc_topic['dominant_topic'] = significant_topic
```

```python
df_doc_topic.head()
```

```python
WTest = lda.transform(vectorizer_tf.transform(X_test.complaints))
```

```python
colnames = ["Topic_" + str(i) for i in range(lda.n_components)]
docnames = ["Doc_" + str(i) for i in range(len(X_test.complaints))]
df_doc_topic = pd.DataFrame(np.round(WTest,2),columns=colnames,index=docnames)
significant_topic = np.argmax(df_doc_topic.values,axis=1)
df_doc_topic['dominant_topic'] = significant_topic
```
